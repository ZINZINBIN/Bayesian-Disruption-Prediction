### Operation regime
- Regime 1: safe regime
- Regime 2: Warning regime or therma quench

```python
import matplotlib.pyplot as plt
import seaborn as sns

disrupt_data = pd.read_csv("./dataset/Bayesian_Disruption_Shot_List_revised.csv")
efit_data = pd.read_pickle("./dataset/Bayesian_Disruption_efit.pkl")

shot_list = disrupt_data.shot.unique()
tag_list = []

efit_data = efit_data[efit_data.shot.isin(shot_list)]

for shot_num in shot_list:
    t_warning = disrupt_data[disrupt_data.shot == shot_num].t_warning.values[0]
    tftsrt = disrupt_data[disrupt_data.shot == shot_num].t_flattop_start.values[0]
    t_tq = disrupt_data[disrupt_data.shot == shot_num].t_tmq.values[0]
    t_cq = disrupt_data[disrupt_data.shot == shot_num].t_ip_min_fault.values[0]

    t = efit_data[efit_data.shot == shot_num].time.values
    
    def func(x):
        if x > t_tq:
            return 'Disruptive'
        elif x < t_tq and x > tftsrt:    
            return 'Normal'
        else:
            return 'Flattop'
    
    tag = efit_data[efit_data.shot == shot_num]['time'].apply(lambda x : func(x)).values
    tag_list.append(tag)

efit_data['tag'] = np.concatenate(tag_list)
```

```python
operational = efit_data[efit_data.tag != 'Flattop'].groupby(['shot','tag']).mean()[['\\nG','\\troyon','\\q95']]
```

```python
sns.scatterplot(x = '\\nG', y = '\\troyon', data = operational, hue = 'tag')
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

operational = efit_data[efit_data.tag != 'Flattop'].groupby(['shot','tag']).mean()[['\\nG','\\troyon','\\q95','\\zxpt2','\\zxpt1']]

tSNE = TSNE(n_components=2, perplexity=64)
v = tSNE.fit_transform(operational.values)

operational['x1'] = v[:,0]
operational['x2'] = v[:,1]

plt.figure(1, figsize = (9,5))
sns.scatterplot(x = 'x1', y = 'x2', data = operational, hue = 'tag')
plt.savefig("./results/analysis_file/tSNE_operation_regime.eps")
```
